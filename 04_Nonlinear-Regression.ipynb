## Nonlinear Regression with Noise
This example demonstrates how to generate a nonlinear function with added noise, and visualize both the true function and the noisy data. This setup can be used as a baseline for fitting polynomial regression models.

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
# Create a range of x values from -5.0 to 5.0 with step 0.1
x = np.arange(-5.0, 5.0, 0.1)

# Define a nonlinear cubic function y = x^3 + x^2 + x
y = 1 * (x**3) + 1 * (x**2) + 1 * (x**1)

# Add Gaussian noise to the data to simulate real-world data
y_noise = 20 * np.random.normal(size=x.size)
y_data = y + y_noise

# Plot the original function (red line)
plt.plot(x, y, 'r')
# Plot the noisy data points (blue circles)
plt.plot(x, y_data, 'bo')
plt.show()
```

```python
def sigmiod(x,beta1,beta2):
    y=1/(1+np.exp(-beta1*(x-beta2)))
    return y
```

```python
beta_1=0.8
beta_2=2000
y_pred=sigmiod(x_data,beta_1,beta_2)
plt.plot(x_data,y_data,'bo')
plt.plot(x_data,y_pred*20000000000000,'r')
plt.show()
```

```python
xdata=x_data/max(x_data)
ydata=y_data/max(y_data)
```

```python
from scipy.optimize import curve_fit
popt, pcov = curve_fit(sigmiod, xdata, ydata)
print(f'B_data1= {popt[0]} ,B_data2= {popt[1]}' )
```

```python
x1=np.linspace(1960,2015,60)
x1=x1/max(x1)
y=sigmiod(x1,*popt)
plt.plot(xdata,ydata,'bo', label='data')
plt.plot(x1,y,'r',linewidth=3, label='pred',)
plt.legend(loc='best')
plt.show()
```
